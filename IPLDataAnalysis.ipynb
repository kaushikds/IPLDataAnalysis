# IPL Data Analysis
This notebook is about practicing the data analysis steps of loading the data from a .csv file into a pandas dataframe and getting some basic information about the dataset using pandas functions:

* Loading .csv file data into a dataframe by using the pandas method - **.read_csv()**
* Printing the first 5 rows of the dataframe by using the pandas method - **.head()**
* Finding out the Number of Rows and the Number of Columns present in the dataframe, or finding the "shape of the data", by printing the dataframe's **shape** attribute
* Finding out the type of data present in each column in the dataframe by printing the dataframe's **dtypes** attribute

```python
# Importing the Pandas library, naming it as pd for shortform, and loading the deliveries.csv file data into a pandas dataframe
import pandas as pd

filename = 'deliveries.csv'

df = pd.read_csv(filename)
```

```python
# Printing the first 5 rows of the dataframe
df.head()
```

```python
# Printing the "shape" of the data, or the number of rows and columns present in the entire dataframe

df.shape
```

```python
# Finding out the data type of each column in the dataframe

df.dtypes
```
